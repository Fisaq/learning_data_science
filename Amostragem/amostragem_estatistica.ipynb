```python
# Função que retorna uma amostra aleatória simples ao receber um dataframe e um tamanho de amostra

def amostragem_aleatoria_simples(dataset, n_amostra):
  amostra = dataset.sample(n = n_amostra)
  return amostra
```

```python
# Função que retorna uma amostra sistemática ao receber um dataframe e um tamanho de amostra

def amostra_sistematica(dataset, amostra):
  intervalo = round(len(df)/amostra)
  rd.seed(1)
  val_init = rd.randint(0, intervalo)
  indices = np.arange(val_init, len(df), step = intervalo)
  amoSis = df.iloc[indices]
  return amoSis
```

**Amostragem por Grupos**

Após a divisão realizamos uma amostra aleatória simples para definir de qual grupo será extraída nossa amostra

```python
# Para isso vamos criar algumas variáveis
grupos = []
id_grupo = 0
cont = 0
```

```python
# Acrescentamos esta nova coluna ao dataset original
df['group'] = grupos
df.head()
```

```python
# Agora selecionamos randomicamente um dos 10 grupos:
rd.seed(1)
rd.randint(0,9)
```

```python
# Logo o grupo da amostra é o sorteado:
amostra_grupo = df['group'] == 2
df[amostra_grupo].head()
```

```python
# Função que retorna uma amostra por grupo ao receber um dataframe e um total de grupos
def amostra_grupos(dataset, qtd_grup):
  grupos = []
  id_grupo = 0
  cont = 0

  intGrup = len(dataset) // qtd_grup

  for _ in dataset.iterrows():
    grupos.append(id_grupo)
    cont +=1
    if cont > intGrup:
      cont = 0
      id_grupo +=1

  dataset['group'] = grupos

  rd.seed(1)
  grupo_sorteado = rd.randint(0, qtd_grup)

  amostra = dataset['group'] == grupo_sorteado
  return dataset[amostra]
```

```python
amostra_grupos(df, 68)
```

**Amostra Estratificada**

```python
# Importando a biblioteca sklearn
from sklearn.model_selection import StratifiedShuffleSplit
```

Para esse tipo de amostragem faremos uma análise da distribuição dos salários entre homens e mulheres

```python
df['income'].value_counts()
```

```python
percent_menor_50 = (7841/len(df)) * 100
print(f'Salários maiores que 50 mil: {percent_menor_50: .2f} %')

percent_maior_50 = (24720/len(df)) * 100
print(f'Salários menores que 50 mil:{percent_maior_50: .2f} %')
```

Como podemos perceber, o percentual de salários menores que 50 mil é muito maior que o de salários maiores que 50 mil. Sendo assim, ao recolhermos uma amostra desses dados, a probabilidade de selecionarmos dados apenas dos salários menores é muito maior. Tendo em vista isso, faz-se necessário a utilização de uma **Amostra Extratificada**.

```python
# Tamanho da base de teste
tam_teste = 100 / len(df)
tam_teste
```

```python
# Para nossa análise vamos dividir a base de dados retirando 10% da mesma para utilizarmos como amostra
split = StratifiedShuffleSplit(test_size=tam_teste)

for x,y in split.split(df, df['income']):
  df_x = df.iloc[x]
  df_y = df.iloc[y]
```

```python
df_x.shape, df_y.shape
```

```python
df_y.head()
```

```python
# Podemos perceber que a amostra extratificada foi realizada com sucesso
# Sendo proporcionalmene igual a forma como o dataset está distribuído
df_y['income'].value_counts()
```

```python
# Função que retorna uma amostra extratificada
def amostra_extratificada(dataset, percentual):
  percentual = percentual/100
  split = StratifiedShuffleSplit(test_size=percentual)

  for _,y in split.split(dataset, dataset['income']):
    y_test = dataset.iloc[y]

  return y_test
```

```python
am = amostra_extratificada(df, 10)
am
```
